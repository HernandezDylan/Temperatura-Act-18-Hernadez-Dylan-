## 1. Fórmulas de Conversión

Para convertir entre estas dos escalas, utilizamos relaciones lineales basadas en los puntos de congelación ($0\text{°C}$ o $32\text{°F}$) y ebullición del agua ($100\text{°C}$ o $212\text{°F}$).

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times 1.8) + 32$$
    *(O también expresado como $\frac{9}{5} \times °C + 32$)*

* **De Fahrenheit a Celsius:**
    $$°C = \frac{°F - 32}{1.8}$$
    *(O también expresado como $\frac{5}{9} \times (°F - 32)$)*

---

## 2. Ejemplos Resueltos (Paso a Paso)

### Ejemplo A: Convertir $25\text{°C}$ (temperatura ambiente) a $\text{°F}$
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Multiplicar:** $25 \times 1.8 = 45$
4.  **Sumar la constante:** $45 + 32 = 77\text{°F}$
**Resultado:** $25\text{°C}$ equivalen a **$77\text{°F}$**.

### Ejemplo B: Convertir $98.6\text{°F}$ (temperatura corporal promedio) a $\text{°C}$
1.  **Identificar la fórmula:** $°C = \frac{°F - 32}{1.8}$
2.  **Sustituir el valor:** $°C = \frac{98.6 - 32}{1.8}$
3.  **Restar la constante:** $98.6 - 32 = 66.6$
4.  **Dividir:** $\frac{66.6}{1.8} = 37\text{°C}$
**Resultado:** $98.6\text{°F}$ equivalen a **$37\text{°C}$**.

### Explicación técnica del código:
* **`float()`**: Usamos esto en el `input` porque la temperatura rara vez es un número entero exacto; permite decimales.
* **`f"{valor:.2f}"`**: Esto sirve para redondear el resultado a solo 2 decimales y que no se vea una cadena infinita de números.
* **`while True`**: Crea el bucle infinito que solo se rompe (`break`) cuando el usuario elige la opción 3.
* **`if __name__ == "__main__":`**: Es una buena práctica en Python para asegurar que el menú se ejecute solo si abres este archivo directamente.

```python
def convertir_C_F():
    celsius = float(input("\nIngresa los grados Celsius a convertir: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertir_F_C():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")

        opcion = input("Selecciona una opción (1-3): ")

        if opcion == "1":
            convertir_C_F()
        elif opcion == "2":
            convertir_F_C()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("¡Hernadez Dylan NL=24")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
